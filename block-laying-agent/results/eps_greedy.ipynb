```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
EPS_START = 0.9
EPS_END = 0.05
EPS_DECAY = -1. * 1000 / np.log((0.99*EPS_START - EPS_END) / (EPS_START - EPS_END))
EPS_DECAY
```

```python
EPS_START = 0.9
EPS_END = 0.05
EPS_DECAY = 93944

steps_done = np.array(range(140000))

eps_threshold = EPS_END + (EPS_START - EPS_END) * np.exp(-1. * steps_done / EPS_DECAY)

eps_threshold[1000] / eps_threshold[0]
```

```python
fig, ax = plt.subplots()

ax.plot(steps_done, eps_threshold)

plt.show()
```
